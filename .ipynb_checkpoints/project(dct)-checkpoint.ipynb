```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv("cancer.csv")
df=df.drop(['id','Unnamed: 32'],axis=1)
df.head()
```

```python
z=['radius_mean','area_mean']
for i in z:
    df[i]=df[i].replace(0,np.NaN)
    mean=int(df[i].mean(skipna=True))
    df[i]=df[i].replace(np.NaN,mean)
    print(df[i])
```

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
iqr=q3-q1
print(iqr)
ub=q3+1.5*iqr
lb=q1-1.5*iqr

median=np.median(df['radius_mean'])
print(median)
```

```python
c=0
r=len(df['radius_mean'])
for j in range (r):
    if(df.iloc[j]['radius_mean']>ub) | (df.iloc[j]['radius_mean']<lb):
        c=c+1
print((c/r)*100)
```
