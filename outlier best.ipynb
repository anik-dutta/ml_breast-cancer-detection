```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("cancer.csv")
print(df)
df=df.drop(['id','Unnamed: 32'],axis=1)
df.head(10)
```

```python
df['radius_mean']=np.log(np.log(df['radius_mean']))
print((df['radius_mean']).skew(axis=0))
sns.distplot(df['radius_mean'],color='purple')
plt.show()
df['radius_mean'].describe()
```

```python
df['texture_mean']=np.sqrt(np.sqrt(np.sqrt(df['texture_mean'])))
print((df['texture_mean']).skew(axis=0))
sns.distplot(df['texture_mean'],color='purple')
plt.show()
df['texture_mean'].describe()
```

```python
df['perimeter_mean']=np.log(np.log(np.log(df['perimeter_mean'])))
print((df['perimeter_mean']).skew(axis=0))
sns.distplot(df['perimeter_mean'],color='purple')
plt.show()
df['perimeter_mean'].describe()
```

```python
df['area_mean']=np.log(np.log(df['area_mean']))
print((df['area_mean']).skew(axis=0))
sns.distplot(df['area_mean'],color='purple')
plt.show()
df['area_mean'].describe()
```

```python
df['smoothness_mean']=np.sqrt(np.sqrt(df['smoothness_mean']))
print((df['smoothness_mean']).skew(axis=0))
sns.distplot(df['smoothness_mean'],color='purple')
plt.show()
df['smoothness_mean'].describe()
```

```python
df['compactness_mean']=np.log(df['compactness_mean'])
print((df['compactness_mean']).skew(axis=0))
sns.distplot(df['compactness_mean'],color='purple')
plt.show()
df['compactness_mean'].describe()
```

```python
zero_not_accepted=['concavity_mean']
for column in zero_not_accepted:
    df[column]=df[column].replace(0,np.NaN)
    median=float(df[column].median(skipna=True))
    df[column]=df[column].replace(np.NaN,median)
df['concavity_mean']=np.sqrt(np.sqrt(df['concavity_mean']))
print((df['concavity_mean']).skew(axis=0))
sns.distplot(df['concavity_mean'],color='purple')
df['concavity_mean'].describe()
```
