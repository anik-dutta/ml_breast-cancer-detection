```python
import numpy as np
import pandas as pd
df=pd.read_csv('Cancer.csv')
df.head(25)
```

```python
f_drop=['radius_worst','radius_mean','radius_se','texture_worst','perimeter_worst','area_worst','smoothness_worst',\
        'compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst',\
        'concavity_se','perimeter_se','concavity_mean','area_mean','compactness_se','concave points_mean']
df=df.drop(f_drop,axis=1) #f_drop: feature drop
```

```python
df.drop(['Unnamed: 32','id',],inplace=True,axis=1)
print(df.head())
```

```python
import seaborn as sns
sns.countplot(df['diagnosis'])
```

```python
df['diagnosis'].value_counts()
```

```python
zero_not_accepted=['texture_mean','perimeter_mean','smoothness_mean','compactness_mean','symmetry_mean',\
                   'fractal_dimension_mean','texture_se','area_se','smoothness_se','concave points_se','symmetry_se','fractal_dimension_se']
for column in zero_not_accepted:
    df[column]=df[column].replace(0,np.NaN)
    mean=int(df[column].mean(skipna=True))
    df[column]=df[column].replace(np.NaN,mean)
    print(df[column])
```

```python
from sklearn.preprocessing import LabelEncoder
le_diagnosis=LabelEncoder()
df['Diagnosis']=le_diagnosis.fit_transform(df['diagnosis'])
df.drop('diagnosis',axis=1,inplace=True)
print(df.head(25))
```

```python
df['Diagnosis'].value_counts()
```

```python
import seaborn as sns
sns.countplot(df['Diagnosis'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
y=df.Diagnosis
x=df.drop('Diagnosis',axis=1)
clf_0=LogisticRegression().fit(x,y)
pred_y_0=clf_0.predict(x)
```

```python
print(accuracy_score(pred_y_0,y))
```

```python
#KNN model
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
#split dataset and into train and test

X=df.iloc[:,0:14]
Y=(df['Diagnosis'])
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=4,test_size=0.3)
```

```python
#Define the model:Init
classifier=KNeighborsClassifier(n_neighbors=14,p=2,metric='euclidean')
classifier.fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_pred))
```
