```python
#outlier detection

import numpy as np
import pandas as pd
outliers=[]
def detect_outlier(cdf):
    threshold=1
    mean_1=np.mean(cdf.radius_mean)
    std_1=np.std(cdf.radius_mean)
    for y in cdf.radius_mean:
            z_score=(y-mean_1)/std_1
            if np.abs(z_score)>threshold:
                outliers.append(y)
    return outliers
    
```

```python
outlier_datapoints=detect_outlier(cdf)
out=len(outlier_datapoints)
print((out/569)*100)
print(outlier_datapoints)
```
